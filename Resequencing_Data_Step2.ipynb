```python
###### STEP2
import pandas as pd

# 读取指定的工作表
file_path = "E:/My_python_work_xibanya/LG1-Step1.xlsx"
sheet_name = "LG1-step1"
df = pd.read_excel(file_path, sheet_name=sheet_name)

# 找到 “A_Count” 列的索引
a_count_index = df.columns.get_loc("A_Count")

# 初始化要删除的行的索引列表
rows_to_drop = []

# 从第1行开始遍历每一行
for index, row in df.iloc[0:].iterrows():
    # 计算 "-" 的数量和总单元格数量（第六列开始到 “A_Count” 列之前）
    total_cells = a_count_index - 6  # 第六列开始
    dash_count = (row[5:a_count_index] == "-").sum()
    
    # 判断 "-" 的数量是否占总数的 20% 或更多
    if dash_count / total_cells >= 0.2:
        # 标记该行为删除
        rows_to_drop.append(index)

# 删除标记的行
df_cleaned = df.drop(rows_to_drop)

# 初始化一个新的列表来存储需要删除的行索引
rows_to_drop = []

# 从第1行开始遍历每一行，检查是否满足删除条件
for index, row in df_cleaned.iloc[0:].iterrows():
    f_a = row['f(A)']
    f_h = row['f(H)']
    
    # 检查是否满足任一删除条件
    if (f_a == 0 and f_h == 0) or (f_a == 1 and f_h == 0) or f_a > 0.9 or f_a < 0.1 or f_h < 0.15:
        rows_to_drop.append(index)

# 删除标记的行
df_final = df_cleaned.drop(rows_to_drop)
from openpyxl import load_workbook
from openpyxl.utils.dataframe import dataframe_to_rows
from openpyxl.styles import PatternFill

# 加载现有的 Excel 文件
file_path = "E:/My_python_work_xibanya/LG1-Step1.xlsx"
wb = load_workbook(file_path)

# 创建一个新的工作表 "Segregating loci"
ws = wb.create_sheet("Segregating loci")

# 将 df_final 数据逐行逐列写入新工作表
for r_idx, row in enumerate(dataframe_to_rows(df_final, index=False, header=True), 1):
    for c_idx, value in enumerate(row, 1):
        cell = ws.cell(row=r_idx, column=c_idx, value=value)
        # 这里可以根据需要设置单元格的格式
        # 例如，根据值设置单元格颜色
        if value == 'A':
            cell.fill = PatternFill(start_color='FFC7CE', end_color='FFC7CE', fill_type='solid')
        elif value == 'B':
            cell.fill = PatternFill(start_color='FFEB9C', end_color='FFEB9C', fill_type='solid')# 浅黄色
        elif value == 'H':
            cell.fill = PatternFill(start_color='C6EFCE', end_color='C6EFCE', fill_type='solid')  
        # 可以根据实际情况添加更多的条件和格式设置

# 保存工作簿
wb.save(file_path)
```
